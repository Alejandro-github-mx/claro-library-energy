```python
import sys
from pathlib import Path

# Start from current working directory and walk upwards until we find a repo marker
here = Path.cwd().resolve()

def find_repo_root(start: Path) -> Path:
    for p in [start] + list(start.parents):
        if (p / ".git").exists() or (p / "pyproject.toml").exists() or (p / "README.md").exists():
            return p
    raise RuntimeError("Could not find repo root (no .git/pyproject.toml/README.md found above).")

repo_root = find_repo_root(here)
sys.path.insert(0, str(repo_root))

print("CWD:", here)
print("Repo root:", repo_root)
```
